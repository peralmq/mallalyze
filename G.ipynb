```python
import collections
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def setup(datafile):
    df = pd.read_csv(datafile, header=0)
    df['end'] = pd.to_datetime(df['end'])
    df['start'] = pd.to_datetime(df['start'])
    df['duration'] = (df['end'] - df['start']).astype('timedelta64[m]')
    df['hour'] = df.start.apply(lambda x: x.hour)
    df['weekday'] = df.start.apply(lambda x: x.weekday())
    df['week'] = df.start.apply(lambda x: x.isocalendar()[1])
    df['day'] = df.start.apply(lambda x: x.day)
    return df

def stats(series):
    return {'mean': series.mean(),
            'std': series.std(),
            'p50': series.median(),
            'p95': series.quantile(0.95),
            'p99': series.quantile(0.99)}
```

```python
df = setup('G.csv')
```

# 1. Duration distribution

```python
df.duration.hist(bins=100)
stats(df.duration)
```

The duration of visits distribution looks like a poisson or gamma distribution. It shows that on average the visitors stays at the mall four about 1-1.5 hours give or take an hour.

# 2. Length of traces distribution

```python
df.events.hist(bins=100)
stats(df.events)
```

The length of traces distribution also looks like a poisson distribution, but with a lower $\lambda$. It shows that on average the traces contain about 6-12 events.

# 3. Unique and frequency of visits

```python
frequency_dist = df.uid.groupby(df.uid.value_counts()).size()
frequency = pd.DataFrame({'frequency': frequency_dist})
frequency.plot()
dict({'total_unique_visitors': len(df.uid.unique()) / float(len(df))}, **stats(frequency_dist))
```

Most visitors visists the mall 2-50 times and the mall has a uniqueness factor of 64%.

## Hourly

```python
total = df.uid.groupby(df['hour']).size()
c = collections.Counter(df.groupby(['hour', 'uid']).size().reset_index(1, drop=True).keys())
unique = pd.Series(c.values(), index=total.index, dtype='int64')
pd.DataFrame({'unique': unique, 'total': total}).plot()
```

Most return visitors (that visits more than once) do it during the day.

## Weekly

```python
total = df.uid.groupby(df['week']).size()
c = collections.Counter(df.groupby(['week', 'uid']).size().reset_index(1, drop=True).keys())
unique = pd.Series(c.values(), index=total.index, dtype='int64')
pd.DataFrame({'unique': unique, 'total': total}).plot(kind='bar')
```

The unique/total ratio doesn't seem to change that much between weeks. (No special deals or drives going on?)

# 4. Model for finding employees

Spending more than 5h at the mall for 2 or more days constitues an employee

```python
working_days = df.groupby(['uid','day'])['duration'].sum()/60 > 5
filter_employees = working_days[working_days == True]
c = collections.Counter(filter_employees.reset_index(1, drop=True).keys())
employees = [uid for uid,count in c.iteritems() if count > 1]
def is_employee(uid):
    return uid in employees
```

```python
is_employee(3), is_employee(526)
```

User 3 is not an employee but 526 is.
